```python
from scipy.stats import norm
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(1, 1)
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
pdf = norm.pdf(x)

ax.plot(x, pdf, 'r-', lw=5, alpha=0.6, label='norm pdf')
ax.plot(x, pdf / sum(pdf), 'b-', lw=5, alpha=0.6, label='norm pdf (normalized)')
```
